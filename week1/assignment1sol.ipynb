```python

file=open("student_records.csv","r")
lines=file.readlines()

gradevalue={
    "AP":10,
    "AA":10,
    "AB":9,
    "BB": 8,
    "BC":7,
    "CC":6,
    "CD":5,
    "DD":4,
    "F": 0,
    "FR":0
}
students={}
for line in lines[1:]:
  line=line.strip()
  list1=line.split(",")
  roll=list1[0]
  credit=int(list1[2])
  grade=list1[5]
  points=gradevalue[grade]
  score=points*credit
  if roll not in students:
    students[roll]={
        'totcredits':0,
        'totscore':0,
    }
  students[roll]['totcredits']+=credit
  students[roll]['totscore']+=score


for roll,value in students.items():
  totcredits=value['totcredits']
  totscore=value['totscore']
  cpi=totscore/totcredits
  print(f"roll={roll} , totalcredits={totcredits} , CPI={cpi:.2f}")

file.close()


```

```python
file=open("student_records.csv","r")
lines=file.readlines()
minorstu={}
def splitnum(coursename):
    for i, char in enumerate(coursename):
        if char.isdigit():
            return coursename[:i]
    return coursename



for line in lines[1:]:
  line=line.strip()
  list1=line.split(",")
  roll=list1[0]
  credit=int(list1[2])
  coursename=list1[3]
  coursetype=list1[4]
  dept=splitnum(coursename)
  if isinstance(dept, tuple):
      dept = dept[0]
  if list1[4]=="minor":
    if roll not in minorstu:
      minorstu[roll]={}
    if dept not in minorstu[roll]:
        minorstu[roll][dept] = 0
    minorstu[roll][dept]+=credit


for roll, departments in minorstu.items():
  for dept, totalcredit in departments.items():
    if totalcredit >= 10:
      print(f"roll:{roll} completed minor in dept:{dept}")

file.close()


```

```python
import pandas as pd
import time
import numpy as np
import matplotlib.pyplot as plt

def load_data(data_path):
    data = pd.read_csv(data_path, header=None).values
    return data

def initialise_centers(data, K, init_centers=None):
    if init_centers is not None:
        return np.array(init_centers)
    idx = np.random.choice(len(data), K, replace=False)
    return data[idx]


def initialise_labels(data):
    return np.ones(len(data), dtype=int)


def calculate_distances(data, centers):
    N = data.shape[0]
    K = centers.shape[0]
    dists = np.zeros((N, K))
    for i in range(K):
        dists[:, i] = np.linalg.norm(data - centers[i], axis=1)
    return dists

def update_labels(distances):
    return np.argmin(distances, axis=1)

def update_centers(data, labels, K):
    centers = []
    for k in range(K):
        cluster_points = data[labels == k]
        if len(cluster_points) == 0:
            centers.append(np.zeros(data.shape[1]))
        else:
            centers.append(np.mean(cluster_points, axis=0))
    return np.array(centers)


def check_termination(labels1, labels2):
    return np.array_equal(labels1, labels2)
def kmeans(data_path:str, K:int, init_centers):

    data = load_data(data_path)
    centers = initialise_centers(data, K, init_centers)
    labels = initialise_labels(data)

    start_time = time.time()

    while True:
        distances = calculate_distances(data, centers)
        labels_new = update_labels(distances)
        centers = update_centers(data, labels_new, K)
        if check_termination(labels, labels_new): break
        else: labels = labels_new

    end_time = time.time()
    return centers, labels, end_time - start_time

def visualise(data_path, labels, centers):
    data = load_data(data_path)
    plt.scatter(data[:, 0], data[:, 1], c=labels, s=50,cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
    plt.title('K-means clustering')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.savefig('kmeans.png')
    return plt
data_path = 'https://raw.githubusercontent.com/viswateja2727/SOC-Visionary/refs/heads/main/Week-1/Assignment/spice_locations.txt'
K, init_centers = 2, None
centers, labels, time_taken = kmeans(data_path, K, init_centers)
print('Time taken for the algorithm to converge:', time_taken)
visualise(data_path, labels,centers)
```
